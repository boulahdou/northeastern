## Functions are foundational to computer science and mathematics.
### A function accepts one or more inputs and returns an output.

```python
# Here is a function that takes in one input x, and outputs its square, x^2.

def f(x):
    return x*x

print(f(-5),f(-1),f(3),f(7))
```

```python
# Here is a function that takes in one input x, and outputs
# x^2+1 if x<-2
# 5 if -2<x<2
# and 2x+1 otherwise

def g(x):
    if x<-2: return x*x+1
    elif -2<=x<2: return 5
    else: return 2*x+1

print(g(-5),g(-1),g(3),g(7))
```

## Let's plot these two functions f(x), g(x)

### To do this, we import three Python libraries called "numpy", "matplotlib", and "math"

```python
%matplotlib inline
from numpy import *
from matplotlib.pyplot import * 
from math import *

x = linspace(-5,5, 101)
y1 = vectorize(f)(x)
y2 = vectorize(g)(x)

plot(x, y1, 'red') 
plot(x, y2, 'blue')
title("Comparing our two functions")
xlabel("x values")
ylabel("y values");
show()
```
